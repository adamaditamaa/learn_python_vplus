```python
import pandas as pd
import vpluslib 

cnn = vpluslib.connect_init(4)

query_fact_payment = '''
select (start_date+interval '7 hour')::date dt,orderid_bss,dim_productid
from  production.fact_payment
where (start_date+interval '7 hour')::date between '2024-10-25' and '2024-10-28'
'''

query_dim_product = '''
select * from production.dim_product
'''

df_fact_payment = pd.read_sql(query_fact_payment,cnn)
df_dim_prod = pd.read_sql(query_dim_product,cnn)





```

```python
df_fact_payment.rename(columns={'dim_productid':'id_sama'},inplace=True)
df_dim_prod.rename(columns={'id':'id_sama'},inplace=True)
```

```python
data_baru = df_fact_payment.merge(df_dim_prod, on='id_sama', how='left')
data_baru
```

```python
data_baru_2 = df_fact_payment.join(df_dim_prod, on='id_sama', how='left', lsuffix='_payment', rsuffix='_product')
```
